Как видно с ростом разницы между реальной и ощущаемой температурой число заказов падает, причина пока не ясна

```python
df['date'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S')
df['hour'] = df['date'].dt.hour
df['month'] = df['date'].dt.month
df.head()
```

```python
df.plot(kind="scatter", x="hour", y="cnt")
```

На графике видно, что число заказов в начале и в конце рабочего дня имеет более широкое распределение

```python
sns.relplot(data=df, x="hour", y="cnt", hue="is_holiday", kind="line")
```

```python
sns.relplot(data=df, x="hour", y="cnt", hue="is_weekend", kind="line")
```

Исходя из графиков можно определить, что в выходные и праздничные дни заказов меньше + они происходят в середине дня, 
в то время как в рабочие дни люди используют велосипеды чтоб добраться на работу

```python
sns.lineplot(x=df.month, y=df.cnt)
```

```python
sns.lineplot(x=df.month, y=df.t1)
```

```python
df_s = df.sample(frac=0.1)
sns.lineplot(x=df_s.t1, y=df_s.cnt,  estimator="mean")
```

Очевидно, число заказов падает с ухудшением погодных условий, однако это падение можно было бы невелировать 
в температурной зоне от 15 до 20 градусов, предложив более выгодные тарифы осенью и весной

```python
from scipy.ndimage.filters import gaussian_filter
```

```python
df_smooth = gaussian_filter(df, sigma=1)
```

```python
ts = pd.Series(df["cnt"].values, index=df["timestamp"])
mean_smoothed = ts.rolling(window=5).mean()
mean_smoothed
```

```python
df2 = pd.DataFrame(mean_smoothed)
df2.columns = ['MA']
df2
```

```python
# df['smoothed_cnt'] = mean_smoothed
# del df[0]
pd.concat([df, df2["MA"]], axis=1).head(15)
```
